## This project will look into specific value counts within a 20 mile radius of Milford NJ, such as jobs, 1 bedroom apartments for rent, homes for sale. Here are the specific questions this project will explore.

- Question 4: What location has the most affordable homes for sale within a 20 mile radius of Milford NJ??

### Define our function for scraping urls to use for each question.

```python
#Get the url to scrape for analysis
def scrape_url(url):
    
    #Use requests to get the url data
    response = requests.get(url)
    
    #Initialize status to check for access
    status = response.status_code
    
    #If statement that uses beautiful soup if cleared to scrape, else print error status
    if status == 200:
        page = response.text
        soup = bs(page)
    else:
        print(f"Oops! Received status code {status}")

    #print the cleaned up html 
    print(soup.prettify())

    #print the type of object the soup variable is
    type(soup)

    return soup
```

```python
#Run the scrape url function to convert into a dataframe later
url1 = scrape_url("https://cnj.craigslist.org/search/jjj?employment_type=1&isTrusted=true&postal=08848&search_distance=20#search=2~thumb~0")
```

### Define our main functions for getting our dataframes and plots

```python

#Function to scrape url with beautiful soup to convert to a dataframe
def get_url(url, url_title, col1, col2):

    #Get the entire list
    table = url.find('ol')

    #Initialize df dictionary list
    df = []

    #Look for every li element in the ordered list
    for row in table.find_all('li'):
        
        #Get all values for first column
        first_col = row.find('div', class_ = col1)
    
        #Get all values for second column
        second_col = row.find('div', class_ = col2)

        #If there are any remaining div's with col1 and col2, run the following
        if first_col and second_col:

            #Capitalize letters
            col2_name = second_col.text.strip().upper()

            #Create a list of dictionaries for col1 and col2
            df.append({url_title: first_col.text.strip().upper(), col2.upper(): col2_name})
    
    dataframe = pd.DataFrame(df)

    return dataframe
```

```python
#Function to plot out the dataframes with a line plot
def get_plot(df, x_title, y_title, plot_div, graph_title):

    #Initialize seaborn graph
    sns.set()

    #Set the style for seaborn
    sns.set_style("whitegrid", {"grid.color": ".7"})
    
    #Set the color palette for readability
    palette = sns.color_palette("pastel")
    
    #Add up the value counts to plot on the y-axis
    y = df[plot_div].value_counts(ascending=True)
        
    #Get the individual unique values to plot on the x-axis
    x = y.index

    #Size of plot
    plt.figure(figsize=(25,10))

    #Rotate x-axis values 90 degrees
    plt.xticks(rotation = 90, fontsize = 20)

    plt.yticks(range(2,len(y),2), fontsize = 20)

    #Plot the x-axis
    plt.xlabel(x_title, fontsize = 40, labelpad = 40)

    #Plot the y-axis
    plt.ylabel(y_title, fontsize = 40, labelpad = 40)

    plt.title(graph_title, fontsize = 60, pad = 40)

    #Plot the graph
    plt.plot(x,y, color='orange')

    #Make a scatter plot for increased readability
    plt.scatter(x,y, marker='o')

    #Remove the top and the right border of the graph
    sns.despine()
```

```python
#Function to plot out the dataframes with a bar graph
def bar_plot(df, x_title, y_title, plot_div, graph_title):

    #Initialize seaborn graph
    sns.set()

    #Set up grid
    sns.set_style("whitegrid", {"grid.color": ".7"})

    #Palette for increased readability
    palette = sns.color_palette("pastel")
    
    #Add up the value counts to plot on the y-axis
    y = df[plot_div].value_counts(ascending=True)
        
    #Get the individual unique values to plot on the x-axis
    x = y.index

    #Size of plot
    plt.figure(figsize=(25,10))

    #Rotate x-axis values 90 degrees
    plt.xticks(rotation = 90, fontsize = 20)

    plt.yticks(range(len(y)), fontsize  =20)

    #Plot the x-axis
    plt.xlabel(x_title, fontsize = 44, labelpad = 40)


    #Plot the y-axis
    plt.ylabel(y_title, fontsize = 44, labelpad = 40)

    plt.title(graph_title, fontsize = 60, pad = 40)

    #Plot the bar graphs
    plt.bar(x,y, color = 'orange')

    #Remove the top and the right border of the graph
    sns.despine()
```

```python
#Function to clean up dataframe
def clean_data(df, col):

    #drop duplicate rows 
    df = df.drop_duplicates()

    #Replace all state abbreviations and typical typos to help filter our dataset and avoid duplicate towns
    df.loc[:, col] = df[col].str.replace('NJ', "")
    df.loc[:, col] = df[col].str.replace('PA', "")
    df.loc[:, col] = df[col].str.strip()
    df.loc[:, col] = df[col].str.strip(',')
    df.loc[:, col] = df[col].str.strip(',')
    df.loc[:, col] = df[col].str.strip(', .')

    #Remove rows with an empty in our defined column
    df = df[df[col] != ""]

    #Return cleaned up dataframe
    return df
```

```python
clean_df_1 = clean_data(question1df, 'LOCATION')

clean_df_1
```
